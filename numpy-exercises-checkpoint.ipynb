## When you try the last cell, it produces an error. Why do think this is?
* The error happens due to a Shape Mismatch. In NumPy, element-wise operations like addition require the arrays to have compatible dimensions.

Array 1 is (5, 3) (5 rows, 3 columns).

Array 2 is (3, 5) (3 rows, 5 columns).

Since the rows and columns don't align, NumPy's Broadcasting rules cannot determine how to pair the individual elements, resulting in a ValueError

## How would you fix it?
* We can fix this by ensuring the shapes match before adding. There are two ways:

Transposing: Use .T to flip the rows and columns of the second array so it becomes (5, 3).

Reshaping: Use .reshape(5, 3) if the total number of elements allows it

## لاWhat does the transpose do?
* Transposing an array (matrix) flips it over its main diagonal. Mathematically, it switches the row and column indices of every element: an element at position $(i, j)$ moves to $(j, i)$.

```python
# Create two arrays of random integers between 0 to 10
# one of size (3, 3) the other of size (3, 2)
arr1=np.random.randint(0,10,size=(3,3))
arr2=np.random.randint(0,10,size=(3,2))
print("arr1=",arr1)
print("arr2=",arr2)
```

# Perform a dot product on the two newest arrays you created
arr3=np.dot(arr1,arr2)
arr3

## It doesn't work. How would you fix it?
* The operation np.dot(arr1, arr2) fails with a ValueError because the shapes are incompatible for matrix multiplication.The Problem: Both arrays have a shape of $(4, 3)$. According to the rules of Linear Algebra, the number of columns in the first matrix must equal the number of rows in the second matrix. Here, $9 \neq 5$.The Fix: We must Transpose the second array to align the inner dimensions. By using arr2.T, its shape becomes $(3, 4)$.The Result: The operation now becomes $(4, 3) \cdot (3, 4)$, which is mathematically valid and results in a $(4, 4)$ output matrix

```python
# Take the latest two arrays, perform a transpose on one of them and then perform 
# a dot product on them both
arr3=arr1.T
arr4=np.dot(arr2,arr3)
arr4
```

## Notice how performing a transpose allows the dot product to happen.

## Why is this?
* According to the rules of Matrix Multiplication, the number of columns in the first matrix must match the number of rows in the second matrix.
When we have two arrays of the same shape, e.g., $(3, 5)$ and $(3, 5)$, they are incompatible for a dot product. By transposing the second array, its shape becomes $(5, 3)$. This aligns the inner dimensions $(3, 5) \cdot (5, 3)$, satisfying the mathematical requirement and resulting in a $(3, 3)$ output matrix

## What happens when you compare the arrays with `>`?
* When using the > operator on two NumPy arrays of the same shape, an element-wise comparison is performed.

Mechanism: NumPy compares each element in the first array with the corresponding element (at the same index) in the second array.

Output: The result is a Boolean Array of the same shape (5, 9).

Values: Each position in the resulting array contains True if the element in arr1 is strictly greater than the element in arr2, and False otherwise.

This is a fundamental concept in NumPy called Vectorization, which allows us to perform logical operations across entire datasets without using slow Python loops.

```python
# Compare the two arrays with '>='
arr1>=arr2
```

```python
# Find which elements of the first array are greater than 
greater_than_7 = arr1[arr1 > 7]
print(greater_than_7)
indices = np.where(arr1 > 7)
print(indices)
```
